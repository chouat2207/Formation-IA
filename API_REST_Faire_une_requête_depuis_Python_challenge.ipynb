```python
import requests
import pandas as pd
```

```python
url = "https://data.nantesmetropole.fr/api/records/1.0/search/?dataset=244400404_stations-velos-libre-service-nantes-metropole-disponibilites&q=&facet=banking&facet=bonus&facet=status&facet=contract_name"
response = requests.get(url)
data = response.json()
```

```python
stations = []
for record in data['records']:
  fields = record['fields']
  address = fields.get("address")
  available_bikes = fields.get("available_bikes", 0)
  bike_stands = fields.get("bike_stands", 1)   # Pour éviter une division par zéro
  fill_rate = available_bikes/bike_stands      # Taux de remplissage

  stations.append([address, fill_rate])

df = pd.DataFrame(stations, columns=["Adresse", "taux_remplissage"])
df
```
