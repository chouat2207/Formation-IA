```python
import json
import requests

link = "https://raw.githubusercontent.com/WildCodeSchool/java-jsonparse/master/weather.json"

# This function gets the file content from the URL
r = requests.get(link)

# You can access to the text content with this function
r.text

data = json.loads( r.text)
```

```python
#Affiche la valeur de l'attribut name
print("City name : ", data['name'])
```

```python
#Affiche les valeurs des attributs lat et lon de l'élément coord contenu dans la racine
print('City latitude :', data['coord']['lat'])
print('City longitude :', data['coord']['lon'])
#data['coord']
```

```python
#Parcours tous les éléments dans weather et affiche le contenu de main (peut-être avec une boucle ?)
print("Weather :", data['weather'][0]['main'])
print("Weather :", data['weather'][1]['main'])
```

```python
import pandas as pd
```

```python
#Essaye maintenant d'importer ton fichier json dans pandas :

# You can directly read JSON from your request object
print(r.json())

# And you can import JSON object to pandas
pd.json_normalize(r.json())
```

record_path="weather" extrait les données de la clé "weather" (qui est une liste de dictionnaires).

```python
df = pd.json_normalize(r.json(), record_path="weather", meta=["name", "visibility"])
df
```

```python
df_filtered = df[["main", "name", "visibility"]]

# Affichage du DataFrame filtré
print(df_filtered)
```
