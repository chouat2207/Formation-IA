```python
phrase = """Les parents affirment avoir d’abord rencontré une conseillère en allaitement, Marie-Ève Sturrock, dans la clinique de la dentiste à Saint-Jean-sur-Richelieu. Celle-ci leur aurait donné « plusieurs arguments » pour les convaincre de procéder à une frénotomie, disent les parents. Comme des avantages « au niveau de l’alimentation quand il allait commencer à manger, au niveau de l’élocution quand il allait commencer à parler », énumère Mme Hardy. Des impacts sur l’estime de soi auraient aussi été évoqués, puisque ces enfants se feraient parfois traiter de « langue de fesses » à l’école, dit Mme Hardy. « Ça semblait être une évidence qu’il fallait aller de l’avant pour faire la frénotomie », ajoute-t-elle.

L’intervention semblait aussi « banale, rapide et sans douleur », dit Beauchemin Aucune anesthésie Juste un appareil laser qui permet de couper le frein de langue avec un jet d’eau. Parmi les rares risques possibles, on aurait évoqué un allaitement peut-être un peu plus difficile dans les deux ou trois premiers jours suivant l’intervention, dit le père. Ou encore, le frein de la langue pourrait se recoller et une nouvelle intervention pourrait devoir être pratiquée. « Ça nous semblait peu de risques pour beaucoup d’avantages pour le reste de sa vie », affirme Beauchemin. Peu après la rencontre avec la conseillère en allaitement, la dentiste Marie-Ève Grégoire aurait procédé à la frénotomie, témoignent les parents. Pour ce faire, Jules aurait été emmailloté et placé dans les bras de son père qui était assis dans la chaise du dentiste, relatent-ils. L’intervention aurait été courte. Quelques minutes tout au plus.

Mais dès qu’on a remis le petit dans les bras de sa mère, celle-ci dit avoir constaté que « ses pleurs étaient différents ». « Je l’ai désemmailloté. Rapidement, il est devenu bleu et mou », dit Mme Hardy. Les parents racontent que des manœuvres de réanimation ont été entamées avant qu’ils ne repartent en ambulance avec Jules pour l’hôpital, où le décès du bébé a été constaté. L’intervention semblait aussi « banale, rapide et sans douleur », dit  Beauchemin. Aucune anesthésie. Juste un appareil laser qui permet de couper le frein de langue avec un jet d’eau. Parmi les rares risques possibles, on aurait évoqué un allaitement peut-être un peu plus difficile dans les deux ou trois premiers jours suivant l’intervention, dit le père. Ou encore, le frein de la langue pourrait se recoller et une nouvelle intervention pourrait devoir être pratiquée. « Ça nous semblait peu de risques pour beaucoup d’avantages pour le reste de sa vie », affirme Beauchemin."""
```

```python
phrase
```

```python
#Transforme ton texte en tokens de phrases :Tokenisation en phrases
sent_token = nltk.sent_tokenize(phrase)
```

```python
sent_token
```

```python
#Transforme ton texte en tokens de mots
word_token = nltk.word_tokenize(phrase)
```

```python
word_token
```

```python
frequency = nltk.FreqDist(word_token)
```

```python
frequency
```

```python
print("Top 10 mots les plus fréquents :")
for word, freq in frequency.most_common(10):
    print(f"{word}: {freq}")
```

```python
import matplotlib.pyplot as plt
```

```python
frequency.plot(10, title="Top 10 mots les plus fréquents")
```

```python
tokens = nltk.word_tokenize(phrase.lower())
tokens_clean = []
for word in tokens:
  if word not in nltk.corpus.stopwords.words():
    tokens_clean.append(word)
tokens_clean
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
words = word_tokenize(phrase)

stop_words = set(stopwords.words('french'))

# Filtrer la ponctuation et les stopwords
cleaned_words = [
    word.lower() for word in words if word.isalpha() and word.lower() not in stop_words
]
```

```python
cleaned_words
```

```python
# Calculer les fréquences des mots nettoyés
freq_dist_cleaned = nltk.FreqDist(cleaned_words)

# Afficher les 10 mots les plus fréquents après nettoyage
print("\nTop 10 mots les plus fréquents après nettoyage :")
for word, freq in freq_dist_cleaned.most_common(10):
    print(f"{word}: {freq}")

# Visualiser les mots les plus fréquents
freq_dist_cleaned.plot(10, title="Top 10 mots fréquents (sans stopwords ni ponctuation)")
```

```python
from nltk.stem import SnowballStemmer
```

```python
stem_en = SnowballStemmer('french')
sent_stem = [stem_en.stem(word) for word in cleaned_words]
sent_stem
```

```python
freq_disc_clean_stemmer = nltk.FreqDist(sent_stem)
```

```python
freq_disc_clean_stemmer
```

Un WordCloud à partir du texte initial, grâce à la méthode generate_from_text


```python
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(width= 480, height=480, max_font_size=200, min_font_size=10)
wordcloud.generate_from_text(phrase)

plt.figure()
plt.imshow(wordcloud , interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

Un WordCloud grâce à la méthode generate_from_frequencies après nettoyage des stopwords et de la ponctuation.


```python
wordcloud = WordCloud(width= 480, height=480, max_font_size=200, min_font_size=10)
wordcloud.generate_from_frequencies(freq_dist_cleaned)

plt.figure()
plt.imshow(wordcloud , interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

Un WordCloud grâce à la méthode generate_from_frequencies après nettoyage des stopwords et de la ponctuation, et application d'un stemmer.
