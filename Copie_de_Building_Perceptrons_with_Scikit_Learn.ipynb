```python
# Load required libraries
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
```

# Diabetese Detection Models

This [dataset](https://raw.githubusercontent.com/mansont/datasets-tests/main/diabetese.csv) contains patient data and their diabetese condition: "1" they have diabetes, "0" they do not have diabetese.


Build the following models and compare their performance:
* A logistic regression model
* A single-layer perceptron model
* A multilayer perceptron

```python
# Your code here
url = "https://raw.githubusercontent.com/mansont/datasets-tests/main/diabetese.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
X = df.drop(columns="diabetes")
y=df['diabetes']
```

```python
X_train, X_test , y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
scaler =StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Logistic regression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
```

```python
y_pred_log_reg = log_reg.predict(X_test)
log_reg_accuracy = accuracy_score(y_test, y_pred_log_reg)
print(f"Logistic Regression Accuracy: {log_reg_accuracy :.2f}")
```

```python
#A single-layer perceptron model
perceptron = Perceptron(max_iter=1000)
perceptron.fit(X_train, y_train)
```

```python
y_pred_perceptron = perceptron.predict(X_test)
perceptron_accuracy = accuracy_score(y_test, y_pred_perceptron)
print(f"Single-Layer Perceptron Accuracy: {perceptron_accuracy:.2f}")
```

```python
#Multilayer Perceptron (MLP) Model
mlp = MLPClassifier(max_iter=1000)
mlp.fit(X_train, y_train)
```

```python
y_pred_mlp = mlp.predict(X_test)
mlp_accuracy = accuracy_score(y_test, y_pred_mlp)
print(f"Multilayer Perceptron Accuracy: {mlp_accuracy:.2f}")
```

### Is there a notable difference in the MLP performance when a ReLU, Sigmoid or SoftMax activation function is used?


```python
# Define the activation functions to compare
activations = ['relu', 'logistic', 'identity']

for activation in activations:
    model = MLPClassifier(hidden_layer_sizes=(10, 10), activation=activation, max_iter=1000, random_state=42)
    model.fit(X_train, y_train)
    y_pred_mlp = model.predict(X_test)
    mlp_accuracy = accuracy_score(y_test, y_pred_mlp)
    print(f"MLP Accuracy with {activation} activation: {mlp_accuracy:.4f}")
```

### Does the network performance change when the density (number of neurons) of the hidden layers change?

```python
# Test with different numbers of neurons in hidden layers
neurons = [(10,), (50,), (100,), (10, 10), (50, 50)]

for layers in neurons:
    model = MLPClassifier(hidden_layer_sizes=layers, max_iter=1000, random_state=42)
    model.fit(X_train, y_train)
    y_pred_mlp = model.predict(X_test)
    mlp_accuracy = accuracy_score(y_test, y_pred_mlp)
    print(f"MLP Accuracy with {layers} neurons: {mlp_accuracy:.4f}")
```

Oui , la performance du réseau change en fonction de la densité des couches cachées.
